```kotlin
%use kandy, dataframe
```

```kotlin
import java.io.File

val input = File("Day02.txt").readText().trim().lines()
input
```

```kotlin
val isIncreasing:(List<Int>) -> Boolean = { it.zipWithNext().all { (a, b) -> (b - a) in 1..3 } }
val isDecreasing:(List<Int>) -> Boolean = { it.zipWithNext().all { (a, b) -> (a - b) in 1..3 }}
```

```kotlin
val df = input.map {
    it.split(" ").map { it.trim().toInt() }
}.map {
    val st = isIncreasing(it) || isDecreasing(it)
    if (st)
        Pair("safe",it)
    else
        Pair("unsafe",it)
}.toDataFrame().rename("first").into("status").rename("second").into("values")

df
```

```kotlin
df.plot {
    bars {
        x("status") {
            axis.name = "Status"
        }
        y("values") {
            axis.name = "Values"
        }

        fillColor = Color.RED
        alpha = 0.85

        borderLine {
            color = Color.WHITE
            width = 1.3
        }
    }
}
```
