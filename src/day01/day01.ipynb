```kotlin
import java.io.File
import kotlin.io.path.Path
import kotlin.io.path.readText

val input = File("Day01.txt").readText().trim().lines()
input
```

```kotlin
val left = input.map { it.substringBefore(" ").toInt() }.sorted()
left
```

```kotlin
val right  = input.map { it.substringAfter(" ").trim().toInt() }.sorted()
right
```

```kotlin
%use kandy, dataframe
```

```kotlin
import org.jetbrains.letsPlot.label.ggtitle

val diff = left.mapIndexed { index, i ->  abs(i-right[index]) }
plot {
    x(left)
    y(right)
    points {
        size(diff)
    }
    layout {
        title = "Total distance - ${diff.sum()}"
    }
}
```
